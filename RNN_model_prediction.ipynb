# Creating LSTM model


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout, LSTM
```

```python
 from sklearn.preprocessing import MinMaxScaler
```

```python
u_b,l_b= np.percentile(X,[1,99])
X=np.clip(X,u_b,l_b)
X.shape
```

```python
X= (X-X.min())/(X.max()-X.min())
Y= (Y-Y.min())/(Y.max()-Y.min())
```

```python
print(Y.min(),Y.max())
```

```python
X_n=np.array(X)
Y_n=np.array(Y)
```

```python
X_n=X_n.reshape(X.shape[0],22,1)
```

```python
Y_n=Y_n.reshape(Y.shape[0],1)
```

```python
print(Y_n.shape)
```

```python
model=Sequential()
model.add(LSTM(400,input_shape=(X_n.shape[1:]),activation='relu',return_sequences=True))



model.add(Dense(512,activation='relu'))

model.add(Dense(256,activation='relu'))

model.add(Dense(128,activation='relu'))

model.add(Dense(64,activation='relu'))

model.add(Dense(32,activation='relu'))

model.add(Dense(16,activation='relu'))

model.add(Dense(8,activation='relu'))

model.add(Dense(1,activation='relu'))

opt=tf.keras.optimizers.Adam(lr=1e-3,decay=1e-4)
model.compile(loss='mean_absolute_error',optimizer=opt,metrics=['mean_absolute_error'])
model.summary()

```

```python
h=model.fit(X_n,Y_n, epochs=100, validation_split=0.25, verbose=2)
```
